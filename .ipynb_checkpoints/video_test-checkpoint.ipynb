```python
import cv2
import skvideo.io
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
videodata = skvideo.io.vread('LandingPlane_180.mp4')
print(videodata.shape)
plt.imshow(videodata[0,:,:,0])
```

```python
videodata_bk = np.mean(videodata, axis = 3)
print(videodata_bk.shape)
plt.imshow(videodata_bk[0,:,:])
```

```python
cap = cv2.VideoCapture('LandingPlane_180.mp4')
if (cap.isOpened()==False):
    print("Error opening Video")
    
frameCount = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
frameWidth = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frameHeight = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))




frames = []

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        
        frame_bk = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        frames.append(frame_bk)
        #cv2.imshow('Frame',frame_bk)
        
        print(frame_bk.shape)
        
        if cv2.waitKey(100) & 0xFF == ord('q'):
            break
            
    else:
        break
        
cap.release()

cv2.destroyAllWindows()
```

```python
frames = np.array(frames)
print(frames.shape)
```
